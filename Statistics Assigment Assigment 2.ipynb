```python
"""
Q1. What are the three measures of central tendency?

The three main measures of central tendency are:

Mean

The average of all data values.

Calculated by adding all values and dividing by the number of values.

Example: Marks = [80, 85, 90] → Mean = (80 + 85 + 90)/3 = 85

Median

The middle value when data is arranged in order.

If the number of values is even, it is the average of the two middle values.

Example: Marks = [80, 85, 90] → Median = 85

Mode

The most frequently occurring value in the dataset.

Example: Marks = [80, 85, 85, 90] → Mode = 85

"""
```

```python
"""
Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?

| Measure    | Definition                                          | How it’s Calculated                                                       | Use / Characteristics                                                                         | Example                                     |
| ---------- | --------------------------------------------------- | ------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------- | ------------------------------------------- |
| **Mean**   | The **average** of all data points                  | Sum of all values ÷ Number of values                                      | - Uses all data points<br>- Sensitive to outliers<br>- Best for **normally distributed data** | Marks = [80, 85, 90] → Mean = 85            |
| **Median** | The **middle value** when data is arranged in order | Sort data → middle value (or average of two middle values if even number) | - Not affected by extreme values<br>- Useful for **skewed data**                              | Marks = [70, 80, 85, 90, 200] → Median = 85 |
| **Mode**   | The **most frequently occurring value**             | Identify the value with highest frequency                                 | - Can have multiple modes<br>- Useful for **categorical data**                                | Marks = [80, 85, 85, 90] → Mode = 85        |

How They Measure Central Tendency

Central tendency indicates the “typical” or “central” value in a dataset.

Each measure gives a slightly different view:

Mean: Average, overall center

Median: Middle, robust to outliers

Mode: Most common value, identifies peaks or frequent occurrences

"""
```

```python
"""
Q4. Find the standard deviation for the given data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

Height Data: [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]
Mean = 177.20625 cm
Standard Deviation (σ) = sqrt( Σ (xi - μ)² / N )
Where:
- xi = each data point
- μ = mean of the data
- N = number of data points
Calculations:
1. Calculate each deviation from the mean, square it, and sum all squared deviations.
    Σ (xi - μ)² = (178 - 177.20625)² + (177 - 177.20625)² + ... + (176.5 - 177.20625)²
                = 0.629 + 0.042 + 1.448 + 0.042 + 0.986 + 0.629 + 4.882 + 3.229 + 7.784 + 4.882 + 2.924 + 0.999 + 0.042 + 21.086 + 0.629 + 0.499
                = 49.926
2. Divide the sum by the number of data points (N = 16).
    Variance (σ²) = 49.926 / 16 = 3.120375
3. Take the square root of the variance to get the standard deviation.
    Standard Deviation (σ) = sqrt(3.120375) ≈ 1.766 cm
Standard Deviation ≈ 1.766 cm   

"""
```

```python
"""
Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

Measures of Dispersion and Their Use

Measures of dispersion tell us how spread out or variable the data is around the center (mean/median).

1. Range
Definition:

The difference between the maximum and minimum values.

What it shows:

Gives a quick idea of how widely the data values vary.

Simple, but affected by extreme values (outliers).

Example:

Data = [10, 12, 14, 15, 18]
Range = 18 − 10 = 8

Interpretation:
Values spread across 8 units.

2. Variance
Definition:

The average of squared differences from the mean.

What it shows:

Measures how far each data point is from the mean.

Large variance → data points are far from the mean → high spread.

Small variance → data points are close to the mean → low spread.

Example:

Suppose the variance is 6.64.
This means that on average, the squared distance from the mean is 6.64.

3. Standard Deviation (SD)
Definition:

Square root of variance (σ = √variance).

What it shows:

Tells how much data points deviate from the mean, in the same units as the data.

Widely used because it is easy to interpret.

Example:

If SD = 2.58, it means:

Most values lie within 2.58 units of the mean.

"""
```

```python
"""
Q8. What do you understand about skewness in data?
Skewness is a measure of the asymmetry of a probability distribution or dataset around its mean. It indicates whether the data is skewed to the left (negatively skewed) or to the right (positively skewed).
- Positive Skewness: When the right tail (higher values) is longer or fatter than the left tail. This indicates that there are a few unusually high values pulling the mean to the right. Example: Income distribution in many countries.
- Negative Skewness: When the left tail (lower values) is longer or fatter than the right tail. This indicates that there are a few unusually low values pulling the mean to the left. Example: Age at retirement in certain professions.
- Zero Skewness: When the data is perfectly symmetrical around the mean, indicating a normal distribution.  

"""
```

```python
"""
Q10. Explain the difference between covariance and correlation. How are these measures used in
statistical analysis?

Covariance and correlation are both measures used to describe the relationship between two variables, but they have some key differences:
Covariance:
- Definition: Covariance measures how two variables change together. It indicates the direction of the linear relationship between variables.
- Calculation: Cov(X, Y) = Σ [(Xi - μX)(Yi - μY)] / N
- Interpretation:   
    - Positive Covariance: Both variables tend to increase or decrease together.
        - Negative Covariance: One variable tends to increase when the other decreases.
        - Zero Covariance: No linear relationship between the variables.    
Correlation:
- Definition: Correlation measures both the strength and direction of the linear relationship between two variables. It is a standardized version of covariance.
- Calculation: Correlation (r) = Cov(X, Y) / (σX * σY)
- Interpretation:   
    - Values range from -1 to +1.
        - +1: Perfect positive linear relationship.
        - -1: Perfect negative linear relationship.
        - 0: No linear relationship.
Usage in Statistical Analysis:
- Covariance is useful for understanding the direction of the relationship but is scale-dependent, making it hard to interpret the strength.
- Correlation provides a standardized measure, making it easier to compare relationships across different datasets or variables.    

"""
```

```python
"""
Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
dataset.

The formula for calculating the sample mean (x̄) is:
x̄ = (Σ xi) / n
Where:      
- xi = each data point
- n = number of data points in the sample
Example Calculation:
Dataset: [10, 15, 20, 25, 30]
1. Calculate the sum of all data points:
   Σ xi = 10 + 15 + 20 + 25 + 30 = 100
2. Count the number of data points:
    n = 5       
3. Apply the formula:
   x̄ = 100 / 5 = 20     
Sample Mean = 20

"""
```
