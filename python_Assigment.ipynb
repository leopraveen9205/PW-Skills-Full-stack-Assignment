```python
# 4
# Creating a list of 10 elements with different data types
my_list = [
    25,                     # int
    3.14,                   # float
    "Hello",                # string
    True,                   # boolean
    None,                   # NoneType
    [1, 2, 3],              # list
    (4, 5, 6),              # tuple
    {"a": 1, "b": 2},       # dictionary
    {7, 8, 9},              # set
    bytes([10, 20, 30])     # bytes
]

for element in my_list:
    print(element, "->", type(element))
```

```python
# 5
A = int(input("Enter number A: "))
B = int(input("Enter number B: "))

count = 0  # to count how many times A is divisible by B

# Check if B is not zero to avoid division error
if B != 0:
    # While A is divisible by B
    while A % B == 0:
        A = A // B   # divide A by B
        count += 1

    if count > 0:
        print(f"Number is divisible. It can be divided {count} times.")
    else:
        print("A is not divisible by B.")
else:
    print("Division by zero is not allowed.")
```

```python
# 6
# Creating a list of 25 integers
numbers = list(range(1, 26))   # [1, 2, 3, ..., 25]

# Checking divisibility by 3
for num in numbers:
    if num % 3 == 0:
        print(num, "is divisible by 3")
    else:
        print(num, "is NOT divisible by 3")
```

```python
# 7
my_list = [1, 2, 3]
print(id(my_list))   # memory address before modification

my_list[0] = 10       # modifying list element
print(id(my_list))   # memory address after modification (same)
```

```python
my_str = "Hello"
print(id(my_str))   # memory address before change

my_str = my_str + " World"  # modifying string
print(id(my_str))   # memory address after change (different)
```
