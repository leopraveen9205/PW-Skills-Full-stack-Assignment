```python
"""
Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in
a given dataset?

Linear Regression works correctly only if certain assumptions are satisfied. These assumptions ensure the model gives reliable and unbiased predictions.

1. Linearity

Assumption:
There should be a linear relationship between independent variables (X) and dependent variable (Y).

Example:
Salary increases linearly with years of experience.

How to Check:

Scatter plot between X and Y

Residual vs Fitted plot

If the points follow a straight-line pattern, the linearity assumption holds.

2. Independence of Errors

Assumption:
The residuals (errors) should be independent of each other. One observation should not influence another.

This is especially important in time series data.

How to Check:

Durbin–Watson Test

Value interpretation:

≈ 2 → No autocorrelation

< 2 → Positive autocorrelation

> 2 → Negative autocorrelation

3. Homoscedasticity

Assumption:
The variance of residuals should be constant across all values of the independent variables.

In simple words, the spread of errors should be uniform.

How to Check:

Residual vs Predicted plot

If the spread is constant, homoscedasticity exists.
If the spread forms a cone/funnel shape, it indicates heteroscedasticity.

4. Normality of Residuals

Assumption:
The residuals should follow a normal distribution.

How to Check:

Histogram of residuals

Q-Q plot

Shapiro-Wilk test

If residuals are approximately bell-shaped, the assumption holds.

5. No Multicollinearity

Assumption:
Independent variables should not be highly correlated with each other.

High correlation between predictors can distort coefficient estimates.

How to Check:

Correlation matrix

Variance Inflation Factor (VIF)

Interpretation of VIF:

VIF = 1 → No correlation

VIF < 5 → Acceptable

VIF > 10 → High multicollinearity problem

| Assumption           | Meaning                            | How to Check             |
| -------------------- | ---------------------------------- | ------------------------ |
| Linearity            | X and Y have linear relationship   | Scatter plot             |
| Independence         | Errors are independent             | Durbin-Watson test       |
| Homoscedasticity     | Constant variance of residuals     | Residual plot            |
| Normality            | Residuals are normally distributed | Histogram / Q-Q plot     |
| No Multicollinearity | Predictors not highly correlated   | VIF / Correlation matrix |


"""
```

```python
"""
Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using
a real-world scenario.
In a linear regression model, the slope and intercept are key parameters that help us understand the relationship between the independent variable(s) and the dependent variable.
 
                  Y=β0​+β1​X
Where:

Y → Dependent variable (target)

X → Independent variable (predictor)

β₀ (Intercept) → Value of Y when X = 0

β₁ (Slope) → Change in Y for a one-unit increase in X

1. Slope (β₁)

Meaning:
The slope indicates how much the dependent variable changes when the independent variable increases by one unit.

Interpretation:
If the slope is positive, Y increases as X increases.
If the slope is negative, Y decreases as X increases.

2. Intercept (β₀)

Meaning:
The intercept represents the value of the dependent variable when the independent variable is zero.

Sometimes this value may not have a practical meaning, but it is necessary for the regression equation.

Real-World Example

Scenario: Predicting salary based on years of experience.

Regression equation:

Salary=30000+5000×(Experience)

Interpretation

Intercept (30000):

When experience = 0 years, the predicted salary is ₹30,000.

This represents the starting salary for someone with no experience.

Slope (5000):

For each additional year of experience, the salary increases by ₹5,000.

Example predictions:

2 years experience → ₹40,000

5 years experience → ₹55,000

"""
```

```python
"""
Q4. Explain the concept of gradient descent. How is it used in machine learning?

Gradient Descent is an optimization algorithm used in machine learning to minimize the loss (error) function by iteratively updating model parameters.

1. Concept of Gradient Descent

In machine learning models (like Linear Regression), the model makes predictions and calculates the error between predicted values and actual values using a loss function (for example, Mean Squared Error).

The goal is to find the best parameters (weights) that minimize this error.

Gradient Descent works by:

Calculating the gradient (slope) of the loss function

Moving the parameters in the opposite direction of the gradient

Repeating the process until the error becomes minimal.

Think of it like finding the bottom of a valley by taking steps downhill.

2. Basic Formula:
θ = θ - α * ∇J(θ)

Where:  
θ → Parameters (weights)
α → Learning rate (step size)
∇J(θ) → Gradient of the loss function with respect to parameters

3. Role of Learning Rate (α)

The learning rate controls how big the step is when updating parameters.

Too small → Training becomes very slow

Too large → Model may overshoot the minimum and not converge

4. Types of Gradient Descent

Batch Gradient Descent

Uses the entire dataset to calculate gradients.

Stable but slow for large datasets.

Stochastic Gradient Descent (SGD)

Updates weights using one training sample at a time.

Faster but noisy updates.

Mini-Batch Gradient Descent

Uses small batches of data.

Most commonly used in deep learning.

5. How It Is Used in Machine Learning

Gradient Descent is used to train models by minimizing the loss function.

Examples:

Linear Regression → minimize Mean Squared Error

Logistic Regression → minimize Log Loss

Neural Networks / Deep Learning → update weights during backpropagation

"""
```
