```python
"""
1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Class

A class is a blueprint or template for creating objects.
It defines:

Attributes (variables/data)

Methods (functions/behaviors)

A class does not occupy memory until an object is created.

Example:

A class Car may define:

Attributes → color, brand, model

Methods → start(), stop()

Object

An object is an instance of a class.
It is created from a class and occupies memory.
Objects have their own data and can use methods defined in the class.

Example:

my_car = Car("Red", "Toyota", "Fortuner")

your_car = Car("Blue", "Honda", "City")

Both are objects belonging to the class Car, but hold different data.


"""
```

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show_details(self):
        print(f"The car is a {self.color} {self.brand}.")

# Creating objects of the class
car1 = Car("Toyota", "Red")
car2 = Car("Honda", "Blue")

# Using object methods
car1.show_details()
car2.show_details()
```

```python
"""
2. Name the four pillars of OOPs.

1. Encapsulation

Wrapping data (variables) and methods (functions) into a single unit (class) and controlling access using access modifiers.

2. Inheritance

One class can acquire the properties and behaviors (methods) of another class.
Example: Child class inherits from Parent class.

3. Polymorphism

Ability of a function or method to behave differently based on the object calling it.
Example: method overriding and method overloading.

4. Abstraction

Showing only essential features while hiding internal details.
Example: You use a car by pressing the accelerator—you don’t see the internal engine workings.

"""
```

```python
"""
3. Explain why the __init__() function is used. Give a suitable example.

The __init__() function in Python is called a constructor.
It is a special method that is automatically executed whenever an object of a class is created.

1. To initialize object attributes

It assigns values to variables (attributes) when the object is created.

2. To set default or initial state of an object

Each object can hold different data because of initialization done inside __init__().

3. To perform setup operations

If any starting configuration is needed, it can be done here.

"""

class Student:
    def __init__(self, name, age):
        self.name = name      # Initializing attributes
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects
s1 = Student("Rahul", 20)
s2 = Student("Anita", 22)

# Displaying details
s1.show()
s2.show()
```

```python
"""
5. What is inheritance? Give an example for each type of inheritance.

In Object-Oriented Programming (OOP), inheritance is a mechanism where one class (child/subclass) acquires the properties and methods of another class (parent/superclass).

It promotes code reusability and hierarchical classification.

"""
```

```python
"""
1. Single Inheritance

A child class inherits from one parent class.
"""

# Parent class
class Parent:
    def greet(self):
        print("Hello from Parent")

# Child class
class Child(Parent):
    def greet_child(self):
        print("Hello from Child")

c = Child()
c.greet()        # Inherited method
c.greet_child()  # Child's own method
```

```python
"""
Multiple Inheritance

A child class inherits from more than one parent class.
"""

class Father:
    def skills_father(self):
        print("Father's skill: Driving")

class Mother:
    def skills_mother(self):
        print("Mother's skill: Cooking")

class Child(Father, Mother):
    def skills_child(self):
        print("Child's skill: Painting")

c = Child()
c.skills_father()
c.skills_mother()
c.skills_child()
```

```python
""" Multilevel Inheritance

A chain of inheritance where class C inherits from B, which inherits from A."""

class Grandparent:
    def message(self):
        print("Message from Grandparent")

class Parent(Grandparent):
    def message_parent(self):
        print("Message from Parent")

class Child(Parent):
    def message_child(self):
        print("Message from Child")

c = Child()
c.message()          # From Grandparent
c.message_parent()   # From Parent
c.message_child()    # From Child
```

```python
"""
Hierarchical Inheritance

Multiple child classes inherit from a single parent class.
"""
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    def greet1(self):
        print("Hello from Child1")

class Child2(Parent):
    def greet2(self):
        print("Hello from Child2")

c1 = Child1()
c2 = Child2()

c1.greet()
c2.greet()
```
