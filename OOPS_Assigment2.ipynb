```python
""" 1. Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle """

# Defining the Vehicle class
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle      # Name of the vehicle
        self.max_speed = max_speed                  # Maximum speed
        self.average_of_vehicle = average_of_vehicle  # Average mileage

    def show_details(self):
        print(f"Vehicle Name: {self.name_of_vehicle}")
        print(f"Maximum Speed: {self.max_speed} km/h")
        print(f"Average: {self.average_of_vehicle} km/l")

# Creating an object of Vehicle class
car = Vehicle("Toyota Fortuner", 180, 15)

# Displaying vehicle details
car.show_details()
```

```python
""" 2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.
"""
# Parent class
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

    def show_details(self):
        print(f"Vehicle Name: {self.name_of_vehicle}")
        print(f"Maximum Speed: {self.max_speed} km/h")
        print(f"Average: {self.average_of_vehicle} km/l")

# Child class
class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} passengers"

# Creating an object of Car
my_car = Car("Toyota Fortuner", 180, 15)

# Display vehicle details
my_car.show_details()

# Display seating capacity
print(my_car.seating_capacity(7))
```

```python
"""
What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
"""

# Parent class 1
class Father:
    def skills_father(self):
        print("Father's skill: Driving")

# Parent class 2
class Mother:
    def skills_mother(self):
        print("Mother's skill: Cooking")

# Child class inheriting from both Father and Mother
class Child(Father, Mother):
    def skills_child(self):
        print("Child's skill: Painting")

# Creating object of Child
c = Child()

# Accessing methods from both parents and child
c.skills_father()   # Method from Father
c.skills_mother()   # Method from Mother
c.skills_child()    # Method from Child
```

```python
"""
4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.
Getter: A method used to access the value of a private attribute.

Setter: A method used to update or modify the value of a private attribute.

In Python, we typically use property() or the @property decorator for getters and @attribute_name.setter for setters.
"""
class Employee:
    def __init__(self, name, salary):
        self.__name = name        # Private attribute
        self.__salary = salary    # Private attribute

    # Getter for name
    @property
    def name(self):
        return self.__name

    # Setter for name
    @name.setter
    def name(self, new_name):
        self.__name = new_name

    # Getter for salary
    @property
    def salary(self):
        return self.__salary

    # Setter for salary
    @salary.setter
    def salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Salary must be positive!")

# Creating an object
emp = Employee("John", 50000)

# Using getter
print("Name:", emp.name)
print("Salary:", emp.salary)

# Using setter
emp.name = "David"
emp.salary = 60000

print("Updated Name:", emp.name)
print("Updated Salary:", emp.salary)

# Trying invalid salary
emp.salary = -1000
```

```python
"""
5. What is method overriding in python? Write a python code to demonstrate method overriding.

Definition:
Method overriding occurs when a child class provides its own implementation of a method that is already defined in the parent class.

The child’s method replaces the parent’s method when called using a child object.

It is used to modify or extend behavior of inherited methods.
"""
# Parent class
class Vehicle:
    def fuel_type(self):
        print("Most vehicles use petrol or diesel.")

# Child class
class ElectricCar(Vehicle):
    # Overriding the fuel_type method
    def fuel_type(self):
        print("Electric cars use electricity as fuel.")

# Creating objects
v = Vehicle()
e = ElectricCar()

# Calling methods
v.fuel_type()   # Parent class method
e.fuel_type()   # Overridden method in child class

```
