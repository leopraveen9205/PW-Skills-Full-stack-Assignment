```python
"""
1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.
"""
import seaborn as sns
import plotly.express as px

# Load titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing values for age and fare to avoid errors
titanic = titanic.dropna(subset=["age", "fare"])

# Plot scatter plot using Plotly Express
fig = px.scatter(
    titanic,
    x="age",
    y="fare",
    color="sex",   # Optional: adds color based on gender
    title="Scatter Plot of Age vs Fare (Titanic Dataset)"
)

fig.show()
```

```python
"""
2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
"""
import plotly.express as px

# Load tips dataset directly from Plotly
tips = px.data.tips()

# Create box plot
fig = px.box(
    tips,
    x="day",        # categorical axis
    y="total_bill", # numeric axis
    color="sex",    # optional: color by gender
    title="Box Plot of Total Bill by Day (Tips Dataset)"
)

fig.show()

```

```python
"""
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

"""
import plotly.express as px

# Load tips dataset from Plotly
tips = px.data.tips()

# Create histogram
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",             # Different colors for each day
    pattern_shape="smoker",  # Different patterns for smoker/non-smoker
    barmode="group",
    title="Histogram of Total Bill by Sex (Colored by Day, Patterned by Smoker)"
)

fig.show()
```

```python
"""
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

"""
import plotly.express as px

# Load iris dataset from Plotly
iris = px.data.iris()

# Create scatter matrix plot
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset"
)

fig.update_traces(diagonal_visible=True)  # show histogram on diagonal
fig.show()
```

```python
"""
5. What is Distplot? Using Plotly express, plot a distplot.

A distplot (distribution plot) is a visualization that shows the distribution of a continuous variable.
It typically includes:

A histogram (frequency of values)

A density curve (KDE) to show the smooth distribution

(Optionally) a rug plot

"""

import plotly.express as px

# Load example dataset
iris = px.data.iris()

# Create a distplot-like histogram + KDE + rug
fig = px.histogram(
    iris,
    x="sepal_length",
    nbins=30,
    histnorm="density",
    marginal="rug",
    opacity=0.7,
    title="Distplot-like Distribution of Sepal Length (Iris Dataset)",
)

fig.show()
```
