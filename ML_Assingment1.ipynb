```python
"""
Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
1. Supervised Learning:
In supervised learning, the model is trained on labeled data, meaning each input has a corresponding output label. The goal is to learn a mapping from inputs to outputs to make predictions on new data.
Example: Email spam detection, where emails are labeled as "spam" or "not spam."    

2. Unsupervised Learning:
In unsupervised learning, the model is trained on unlabeled data, meaning there are no output labels provided. The goal is to discover hidden patterns or structures in the data.
Example: Customer segmentation, where customers are grouped based on their purchasing behavior without predefined categories.

3. Semi-Supervised Learning:
In semi-supervised learning, the model is trained on a combination of labeled and unlabeled data. This approach leverages both types of data to improve learning performance.
Example: Image classification where only a small portion of images are labeled, and the rest are unlabeled.
The main differences can be summarized as follows:
- Labeled Data: Supervised learning uses fully labeled data, unsupervised learning uses unlabeled data, and semi-supervised learning uses a mix of both.
- Learning Objective: Supervised learning aims to predict outputs, unsupervised learning aims to find patterns, and semi-supervised learning aims to improve predictions using limited labels.      

"""
```

```python
"""
Q6: What is train, test and validation split? Explain the importance of each term.
In machine learning, the dataset is typically divided into three subsets: training set, validation set, and test set. This division is crucial for building and evaluating models effectively.  
1. Training Set:
The training set is the portion of the dataset used to train the machine learning model. The model  
learns patterns and relationships from this data to make predictions. It is essential for the model to generalize well to new, unseen data.
2. Validation Set:
The validation set is a subset of the training data used to tune hyperparameters and evaluate the model's performance during training. It helps in selecting the best model configuration and prevents overfitting.
3. Test Set:
The test set is a separate portion of the dataset used to evaluate the final model's performance after training and validation. It provides an unbiased assessment of how well the model generalizes to unseen data.            
Each of these splits plays a vital role in the machine learning workflow:
- Training Set: Enables the model to learn from data.
- Validation Set: Helps in model selection and hyperparameter tuning.
- Test Set: Provides an unbiased evaluation of the model's performance on new data.
    
"""
```

```python
"""
Q7: How can unsupervised learning be used in anomaly detection?
Unsupervised learning can be effectively used in anomaly detection by identifying patterns and structures in the data without prior knowledge of what constitutes normal or anomalous behavior. Here are some common approaches:
1. Clustering:  
    Unsupervised algorithms like K-Means or DBSCAN can group similar data points together. Anomalies are identified as points that do not belong to any cluster or are located far from cluster centroids.  
2. Density-Based Methods:
    Techniques like Local Outlier Factor (LOF) or Isolation Forest analyze the density of data points. Anomalies are detected as points in low-density regions compared to their neighbors.
3. Autoencoders:
    Autoencoders are neural networks trained to reconstruct input data. Anomalies can be detected by measuring the reconstruction error; data points with high errors are considered anomalies. 
4. Dimensionality Reduction:
    Methods like Principal Component Analysis (PCA) can reduce the data's dimensionality, making it easier to identify outliers that deviate significantly from the main data distribution. 
    By leveraging these unsupervised learning techniques, organizations can effectively identify unusual patterns or behaviors in various applications such as fraud detection, network security, and fault detection in manufacturing processes.       

"""
```
