```python
"""
Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].
Did you observe any difference in both the outputs? If so then explain it.
"""
####### 1. Reindex the DataFrame
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

df = pd.DataFrame({'course_name': course_name, 'duration': duration})

reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)
```

```python
"""
Step 2: Outputs
Output of new_df.loc[2]

loc works based on label, so it will return the row whose index label is 2.

That row is:

| course_name | Big Data |
| duration | 6 |

Output of new_df.iloc[2]

iloc works based on integer position.
Index position 2 corresponds to the third row in the reordered DataFrame:

Difference Observed

Yes, there is a clear difference.

✔ loc → returns the row with index label 2

That is: Big Data, duration 6

✔ iloc → returns the row at position 2

That is: Machine Learning, duration 3

Why does this happen?

reindex() keeps the original labels, but rearranges their order.

Therefore:

loc looks at the index label, not position.

iloc looks at the position, not the label.

Simple Memory Trick

loc → label

iloc → index location (position)

"""
```

```python
"""
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

Window functions in pandas are operations that perform calculations across a sliding window of data.
These windows move across the dataset to compute statistics such as:

moving averages

cumulative sums

rolling statistics (mean, std, min, max…)

exponentially weighted metrics

They allow us to analyze trends over time or sequence while preserving the original data length.

Pandas provides three major types of window operations:

1️. Rolling Window (df.rolling())

Used to calculate statistics over a fixed-size moving window.

Example metrics:
mean
sum
min
max
std
median

"""
df['values'].rolling(window=3).mean()
```

```python
"""
2. Expanding Window (df.expanding())
Calculates metrics over an expanding dataset starting from the first row.
The window grows with each step.
Example:
cumulative mean
cumulative sum
cumulative standard deviation
"""
df['values'].expanding().mean()
```

```python
""" 3. Exponentially Weighted Window (df.ewm())
Applies exponential weights, giving more importance to recent observations.
Useful for:
smoothing noisy data
financial time series analysis
Example metrics:
ewm.mean()
ewm.std()
ewm.var()

| Window Function                 | Meaning                                         | Use Case                                    |
| ------------------------------- | ----------------------------------------------- | ------------------------------------------- |
| **Rolling**                     | Fixed-size moving window                        | Moving averages, moving std, trend analysis |
| **Expanding**                   | Increasing window (from start to current row)   | Cumulative stats                            |
| **EW (Exponentially Weighted)** | Weighted window with more weight to recent data | Smoothing, financial signals                |


"""
df['values'].ewm(span=3).mean()
```

```python
"""
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) 
and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.
"""
import pandas as pd

# Taking user input
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Converting to pandas datetime
d1 = pd.to_datetime(date1)
d2 = pd.to_datetime(date2)

# Calculating difference
delta = abs(d2 - d1)

# Extracting days, hours, minutes
days = delta.days
hours = delta.seconds // 3600
minutes = (delta.seconds % 3600) // 60

# Displaying results
print(f"Difference between the two dates:")
print(f"Days: {days}")
print(f"Hours: {hours}")
print(f"Minutes: {minutes}")
```

```python
"""
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.
"""
import pandas as pd

# Prompt user for file path
file_path = input("Enter the CSV file path: ")

# Read the CSV file
df = pd.read_csv(file_path)
print("\nFile loaded successfully!\n")

# Show available columns
print("Available columns in the file:")
print(df.columns.tolist())

# Prompt user for column name
column_name = input("\nEnter the column name to convert into categorical: ")

# Prompt user for category order
print("\nEnter the category order separated by commas (e.g., High,Medium,Low): ")
category_order = input().split(",")

# Strip extra spaces
category_order = [cat.strip() for cat in category_order]

# Convert to categorical type with order
df[column_name] = pd.Categorical(df[column_name],
                                 categories=category_order,
                                 ordered=True)

# Sort dataframe based on the categorical column
sorted_df = df.sort_values(column_name)

# Display the sorted data
print("\nSorted DataFrame:")
print(sorted_df)
```

```python
"""
Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

"""
import pandas as pd
import matplotlib.pyplot as plt

# Prompt user for the CSV file path
file_path = input("Enter the CSV file path: ")

# Read the CSV file
df = pd.read_csv(file_path)

print("\nFile loaded successfully!\n")

# Convert the Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index (optional but useful)
df.set_index('Date', inplace=True)

# Identify product category columns (all columns except Date)
# Since Date is index now, all remaining columns are product categories
product_columns = df.columns.tolist()

print("Product categories detected:", product_columns)

# Plotting stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title("Sales of Product Categories Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
"""
Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
"""
import pandas as pd

# Prompt the user to enter the CSV file path
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the "Test Score" column
scores = df['Test Score']

# Calculate mean, median, mode
mean_val = scores.mean()
median_val = scores.median()
mode_val = scores.mode().tolist()   # mode may have multiple values

# Convert mode list to comma-separated string
mode_str = ", ".join(map(str, mode_val))

# Create a results table using a DataFrame
results = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value": [mean_val, median_val, mode_str]
})

# Display the results in a table
print("\n+-----------+----------------+")
print("| Statistic | Value          |")
print("+-----------+----------------+")

for stat, val in zip(results["Statistic"], results["Value"]):
    print(f"| {stat:<9} | {str(val):<14} |")

print("+-----------+----------------+")
```
