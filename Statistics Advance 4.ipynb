```python
"""
Q4: Explain Bayes's theorem with an example.

A4: Bayes's theorem is a mathematical formula used to determine the conditional probability of an event based on prior knowledge of related events. It is expressed as: P(A|B) = [P(B|A) * P(A)] / P(B)     
where P(A|B) is the probability of event A occurring given that event B has occurred, P(B|A) is the probability of event B occurring given that event A has occurred, P(A) is the prior probability of event A, and P(B) is the prior probability of event B.
For example, suppose a medical test is used to detect a disease that affects 1% of the population. The test has a 99% sensitivity (true positive rate) and a 95% specificity (true negative rate). If a person tests positive, we can use Bayes's theorem to calculate the probability that they actually have the disease. 
Let A be the event that the person has the disease, and B be the event that the person tests positive. We know:
P(A) = 0.01 (prior probability of having the disease)       
P(B|A) = 0.99 (probability of testing positive given having the disease)
P(B|¬A) = 0.05 (probability of testing positive given not having the disease)
Using Bayes's theorem, we can calculate P(A|B): 
P(B) = P(B|A) * P(A) + P(B|¬A) * P(¬A)
     = (0.99 * 0.01) + (0.05 * 0.99)
     = 0.0099 + 0.0495
     = 0.0594       
P(A|B) = [P(B|A) * P(A)] / P(B)
       = (0.99 * 0.01) / 0.0594
       ≈ 0.1667 
         Therefore, if a person tests positive, there is approximately a 16.67% chance that they actually have the disease, despite the high sensitivity and specificity of the test.   

Difference Between Bayes's Theorem and Conditional Probability
1. Definition           
Bayes's Theorem: A formula that calculates the conditional probability of an event based on prior knowledge of related events.
Conditional Probability: The probability of an event occurring given that another event has occurred.
2. Formula
Bayes's Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)
Conditional Probability: P(A|B) = P(A ∩ B) / P(B)       
3. Application
Bayes's Theorem: Used to update the probability of an event based on new evidence or information.
Conditional Probability: Used to determine the likelihood of an event given the occurrence of another event.


"""
```

```python
"""
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.

A11: To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can use the following formula for the confidence interval for a population proportion:
Confidence Interval = p̂ ± Z * √(p̂(1 - p̂) /  n)
Where:      
- p̂ is the sample proportion (0.65)
- Z is the critical value for the desired confidence level (for 95% confidence, Z ≈ 1.96)
- n is the sample size (500)    
First, we need to calculate the standard error (SE) of the proportion:
SE = √(p̂(1 - p̂) / n)        
    = √(0.65 * (1 - 0.65) / 500)     
    = √(0.65 * 0.35 / 500)     
    = √(0.2275 / 500)     
    = √0.000455     
    ≈ 0.0213    

Now, we can calculate the margin of error (ME):
ME = Z * SE 
    = 1.96 * 0.0213     
    ≈ 0.0417      
    Finally, we can calculate the confidence interval:
Confidence Interval = p̂ ± ME
                   = 0.65 ± 0.0417
                   ≈ (0.6083, 0.6917)   
Therefore, we can say with 95% confidence that the true proportion of people who are satisfied with their job falls within the interval (0.6083, 0.6917). This means that if we were to take many samples and calculate the confidence interval for each, approximately 95% of those intervals would contain the true population proportion of job satisfaction.    
Difference Between Proportion and Percentage
1. Definition   
Proportion: A ratio that represents a part of the whole, expressed as a decimal between 0 and 1.
Percentage: A ratio that represents a part of the whole, expressed as a fraction of 100.
2. Calculation
Proportion: Calculated as the number of favorable outcomes divided by the total number of outcomes (p̂ = x / n).
Percentage: Calculated as the proportion multiplied by 100 (Percentage = p̂ * 100).
3. Usage    
Proportion: Used in statistical calculations and analyses to represent relative frequencies.
Percentage: Used in everyday language and reporting to express relative frequencies in a more understandable format (e.g., 65% instead of 0.65).


"""
```
