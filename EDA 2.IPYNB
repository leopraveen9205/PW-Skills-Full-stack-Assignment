```python
"""
Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have?
A1. To determine the dimensions of the flight price dataset, you would typically load the dataset using a library like pandas in Python and then use the .shape attribute to get the number of rows and columns. For example:
"""
import pandas as pd 
df = pd.read_csv('flight_price_dataset.csv')
dimensions = df.shape
print(f"The dataset has {dimensions[0]} rows and {dimensions[1]} columns.")
"""
The exact number of rows and columns will depend on the specific dataset being used.
"""
```

```python
"""
Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.
A2. To visualize the distribution of flight prices in the dataset, you can use the matplotlib or seaborn library to create a histogram. Here is an example using matplotlib:
"""
import matplotlib.pyplot as plt
plt.hist(df['Price'], bins=30, edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
"""
This code will create a histogram that shows the frequency of different flight prices in the dataset, allowing you to see the distribution of prices.
"""
```

```python
"""
Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

A3. To find the range of prices in the dataset, you can use the .min() and .max() functions on the 'Price' column of the DataFrame. Here is an example:"""
min_price = df['Price'].min()   
max_price = df['Price'].max()
print(f"The minimum price is {min_price}")
print(f"The maximum price is {max_price}")
```

```python
"""
Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.

A4. To compare the prices of different airlines using a boxplot, you can use the seaborn library in Python. Here is an example:"""
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')   
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
