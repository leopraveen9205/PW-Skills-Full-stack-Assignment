```python
"""
What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame is a 2-dimensional labeled data structure in Pandas, similar to a table in a database or an Excel spreadsheet.
It consists of rows and columns, where each column can have a different data type (e.g., integer, string, float).

Key Features of a DataFrame:

Tabular data with rows and columns.

Each column is a Pandas Series.

Columns can have different data types.

Supports indexing, slicing, filtering, and many operations.

What is a Series in Pandas?

A Series is a 1-dimensional labeled array that can hold any data type (integer, string, float, etc.).
It is like a single column of a DataFrame.

| Feature    | Pandas Series          | Pandas DataFrame                     |
| ---------- | ---------------------- | ------------------------------------ |
| Dimensions | 1-dimensional          | 2-dimensional                        |
| Structure  | Single column          | Multiple columns                     |
| Data type  | All elements same type | Each column can have different types |
| Indexing   | Single index           | Row index and column labels          |

"""
import pandas as pd

# Creating a Series
ages = pd.Series([25, 30, 27])
print("Series:\n", ages)

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)

```

```python
"""
5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

| Function/Method                 | Purpose                                                                                       |
| ------------------------------- | --------------------------------------------------------------------------------------------- |
| `head(n)`                       | Shows the first `n` rows of the DataFrame.                                                    |
| `tail(n)`                       | Shows the last `n` rows of the DataFrame.                                                     |
| `info()`                        | Provides summary information about the DataFrame, including column types and non-null counts. |
| `describe()`                    | Gives statistical summary (mean, std, min, max, quartiles) of numeric columns.                |
| `shape`                         | Returns the number of rows and columns `(rows, columns)`.                                     |
| `columns`                       | Returns a list of column names.                                                               |
| `dtypes`                        | Shows the data type of each column.                                                           |
| `sort_values(by='col')`         | Sorts the DataFrame by a specific column.                                                     |
| `drop(columns=['col'])`         | Removes specified columns.                                                                    |
| `rename(columns={'old':'new'})` | Renames columns.                                                                              |
| `filter()`                      | Selects specific columns or rows based on a condition.                                        |
| `groupby()`                     | Groups data based on column(s) and allows aggregation.                                        |
| `fillna()` / `dropna()`         | Handles missing data.                                                                         |
| `apply()`                       | Applies a function to each row or column.                                                     |
| `loc[]` / `iloc[]`              | Access specific rows/columns by label or index.                                               |

"""

import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Claire', 'David'],
    'Math': [85, 92, 78, 90],
    'Science': [88, 95, 80, 85]
}

df = pd.DataFrame(data)

# Example: Get the average score for Math using describe()
print(df['Math'].describe())

# Example: Sort the DataFrame by Science scores
sorted_df = df.sort_values(by='Science', ascending=False)
print(sorted_df)
```

```python
"""
Create a DataFrame using multiple Series. Explain with an example.

Explanation:

Each Series represents a column of data.

We pass the Series as values in a dictionary to pd.DataFrame(), where the keys become column names.

The resulting DataFrame is 2-dimensional, combining all Series into a tabular structure.
"""
import pandas as pd

# Create individual Series
names = pd.Series(['Alice', 'Bob', 'Claire'])
ages = pd.Series([25, 30, 27])
genders = pd.Series(['Female', 'Male', 'Female'])

# Combine Series into a DataFrame
df = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'Gender': genders
})

# Print the DataFrame
print(df)

```
