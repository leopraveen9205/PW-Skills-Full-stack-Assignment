```python
"""
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation Statistics is a branch of statistics that focuses on estimating population parameters based on sample data. It involves using sample statistics to make inferences about the characteristics of a larger population. There are two main types of estimates in Estimation Statistics: point estimates and interval estimates.         
A point estimate is a single value that serves as an estimate of a population parameter. It is calculated from the sample data and provides a specific value that represents the best guess of the parameter. For example, the sample mean (average) is often used as a point estimate for the population mean.
An interval estimate, on the other hand, provides a range of values within which the population parameter is likely to fall. This range is typically expressed as a confidence interval, which includes a lower and upper bound. The confidence level (e.g., 95%) indicates the degree of certainty that the interval contains the true population parameter. 
Interval estimates account for the variability in the sample data and provide a more comprehensive understanding of the parameter being estimated.    

"""
```

```python
"""
Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
"""
import math
from scipy import stats

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence=0.95):
    """
    Estimate the population mean using the sample mean, sample standard deviation, and sample size.
    
    Parameters:
    sample_mean : float - mean of the sample
    sample_std  : float - standard deviation of the sample
    sample_size : int   - number of observations in the sample
    confidence  : float - confidence level (default 0.95)
    
    Returns:
    tuple: (lower_bound, upper_bound) of the confidence interval for population mean
    """
    # Degrees of freedom
    df = sample_size - 1
    
    # t critical value for confidence level
    t_crit = stats.t.ppf((1 + confidence) / 2, df)
    
    # Margin of error
    margin_error = t_crit * (sample_std / math.sqrt(sample_size))
    
    # Confidence interval
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    
    return lower_bound, upper_bound

# Example usage:
sample_mean = 50
sample_std = 10
sample_size = 25

ci = estimate_population_mean(sample_mean, sample_std, sample_size)
print(f"Estimated population mean is between {ci[0]:.2f} and {ci[1]:.2f}")
```

```python
"""
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

"""
import numpy as np
from scipy.stats import ttest_ind

# Sample data from two populations
# Example: weights of male and female students
male_weights = np.array([72, 75, 78, 80, 74, 77, 79, 81])
female_weights = np.array([60, 62, 65, 63, 64, 67, 66, 61])

# Conduct two-sample independent t-test
# equal_var=False assumes unequal variances (Welch's t-test)
t_stat, p_value = ttest_ind(male_weights, female_weights, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Significance level
alpha = 0.05

# Hypothesis:
# H0: μ_m = μ_f  (no difference)
# H1: μ_m ≠ μ_f  (means are different)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two means.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

```python
"""
Q7: Write down the steps involved in hypothesis testing.

Steps involved in Hypothesis Testing:   
1. Formulate Hypotheses:  
   - Null Hypothesis (H₀): A statement of no effect or no difference.  
   - Alternative Hypothesis (H₁ or Ha): A statement that contradicts H₀.        
2. Choose Significance Level (α):  
   - Common choices are 0.05, 0.01, or 0.10, representing the probability of rejecting H₀ when it is true.  
3. Collect Data:    
    - Gather a representative sample from the population.   
4. Select the Appropriate Test:  
   - Choose a statistical test based on the data type and research question (e.g., t-test, chi-square test).  
5. Calculate Test Statistic:
    - Compute the test statistic using the sample data.
6. Determine the P-value or Critical Value:  
   - P-value: The probability of observing the data if H₀ is true.  
    - Critical Value: The threshold value that the test statistic must exceed to reject H₀.
7. Make a Decision:  
   - If P-value ≤ α, reject H₀; otherwise, fail to reject H₀.  
   - Alternatively, if the test statistic exceeds the critical value, reject H₀.    
8. Draw Conclusions:  
   - Interpret the results in the context of the research question. 


"""
```

```python
"""
Q8. Define p-value and explain its significance in hypothesis testing.
The p-value, or probability value, is a statistical measure that helps determine the significance of the results obtained from a hypothesis test. It represents the probability of observing the data, or something more extreme, assuming that the null hypothesis (H₀) is true. In other words, the p-value quantifies how likely it is to obtain the observed results purely by chance under the assumption that there is no real effect or difference in the population.        

Significance of p-value in Hypothesis Testing:
1. Decision Making: The p-value is used to decide whether to reject or fail to reject the null hypothesis. A smaller p-value indicates stronger evidence against H₀.
2. Significance Level Comparison: The p-value is compared to a predetermined significance level (α, often set at 0.05). If the p-value is less than or equal to α, it suggests that the observed data is unlikely under H₀, leading to its rejection.       

"""
```

```python
"""
Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.
"""
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import t
# Degrees of freedom
df = 10
# x values
x = np.linspace(-4, 4, 1000)
# t-distribution PDF
y = t.pdf(x, df)    
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"Student's t-Distribution (df={df})", color='blue')
plt.fill_between(x, y, where=(x >= -2) & (x <= 2), color='lightblue', alpha=0.5)
plt.title("Student's t-Distribution with 10 Degrees of Freedom")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()

```

```python
"""
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
"""
import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples of equal size (example)
np.random.seed(42)  # for reproducibility
sample1 = np.random.normal(loc=50, scale=5, size=30)  # mean=50, sd=5
sample2 = np.random.normal(loc=52, scale=5, size=30)  # mean=52, sd=5

# Conduct two-sample t-test (independent samples)
# equal_var=True is valid since sample sizes equal and variances assumed equal
t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=True)

print("Sample 1:", sample1)
print("Sample 2:", sample2)
print("\nT-statistic:", t_statistic)
print("P-value:", p_value)

# Null Hypothesis H0: μ1 = μ2
alpha = 0.05

if p_value < alpha:
    print("Reject H0: Significant difference between the two population means.")
else:
    print("Fail to reject H0: No significant difference between the means.")
```

```python
"""
Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic is a standardized value that measures the difference between a sample mean and a population mean (or between two sample means) in terms of standard errors. It is used in hypothesis testing to determine whether to reject the null hypothesis. The t-statistic helps assess how far the sample mean is from the population mean, relative to the variability in the data. A higher absolute value of the t-statistic indicates a greater difference between the means, which may suggest that the observed effect is statistically significant.   
The formula for the t-statistic depends on the context, but for a one-sample t-test, it is given by:
t = (X̄ - μ) / (s / √n)
Where:
- X̄ = Sample mean
- μ = Population mean (under the null hypothesis)
- s = Sample standard deviation
- n = Sample size   
For a two-sample t-test (independent samples), the formula is:
t = (X̄1 - X̄2) / √((s1²/n1) + (s2²/n2))
Where:  
- X̄1, X̄2 = Sample means of the two groups
- s1, s2 = Sample standard deviations of the two groups 


"""
```
