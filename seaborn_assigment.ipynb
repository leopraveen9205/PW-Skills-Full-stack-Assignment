```python
"""
2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Here is the Python code to load the fmri dataset from Seaborn and plot a line plot using:

x = "timepoint"

y = "signal"

Different events

Different regions
"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load fmri dataset
fmri = sns.load_dataset("fmri")

# Create line plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    style="region"
)

plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
""" 3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load titanic dataset
titanic = sns.load_dataset("titanic")

# Plot 1: Box plot of Age vs Passenger Class
plt.figure(figsize=(10, 5))
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

# Plot 2: Box plot of Fare vs Passenger Class
plt.figure(figsize=(10, 5))
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
"""
4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram of price with cut as hue
plt.figure(figsize=(10, 6))
sns.histplot(
    data=diamonds,
    x="price",
    hue="cut",
    kde=False
)

plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
"""
5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

"""

import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Pair plot with species as hue
sns.pairplot(data=iris, hue="species")

plt.show()
```

```python
"""
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix format for heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")

plt.title("Heatmap of Flight Passengers Data")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
