```python
"""
1. Is there any difference in the data type of variables list_ and array_list? 
If there is then write a code to print the data types of both the variables.

Yes, there is a difference in the data type of the variables list_ and array_list.

list_ is a Python list.

array_list is a NumPy ndarray.

Even though both store the same values, their data types are different. """
import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

print("Type of list_:", type(list_))
print("Type of array_list:", type(array_list))
```

```python
"""2. Write a code to print the data type of each and every element of both the variables list_ and
arra_list. """

import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

# Printing data type of each element in list_
print("Data types of elements in list_:")
for element in list_:
    print(element, "->", type(element))

# Printing data type of each element in array_list
print("\nData types of elements in array_list:")
for element in array_list:
    print(element, "->", type(element))
```

```python
"""
3. Considering the following changes in the variable, array_list:

In list_, all elements are strings (str), because they are enclosed in quotes.

When you create array_list = np.array(list_), NumPy infers the datatype â€” since all elements are strings, the dtype becomes <U1 (Unicode string).

When you create array_list = np.array(list_, dtype=int), NumPy converts each string into an integer.
So the elements become integers (int), and the dtype becomes int64 (on most systems).

"""

import numpy as np

list_ = ['1', '2', '3', '4', '5']

# Case 1: Normal numpy array from list of strings
array_list_str = np.array(list_)

# Case 2: Numpy array with dtype=int
array_list_int = np.array(list_, dtype=int)

print("Data types in list_:")
for i in list_:
    print(i, type(i))

print("\nData types in array_list_str:")
for i in array_list_str:
    print(i, type(i))

print("\nData types in array_list_int:")
for i in array_list_int:
    print(i, type(i))
```
