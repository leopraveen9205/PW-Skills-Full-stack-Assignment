```python
"""
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

A1. To calculate the Pearson correlation coefficient (r) between the amount of time students spend studying and their final exam scores, we can use the formula:
r = Σ((X - mean(X)) * (Y - mean(Y))) / (sqrt(Σ(X - mean(X))^2) * sqrt(Σ(Y - mean(Y))^2))
Where:
- X represents the study time
- Y represents the exam scores
- mean(X) is the average study time
- mean(Y) is the average exam score 
Interpretation:
- If r is close to +1, it indicates a strong positive linear relationship, meaning that as study time increases, exam scores tend to increase.
- If r is close to -1, it indicates a strong negative linear relationship, meaning that as study time increases, exam scores tend to decrease.
- If r is close to 0, it indicates no linear relationship between the two variables.    


"""
```

```python
"""
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

A2. To calculate Spearman's rank correlation coefficient (ρ) between the amount of sleep and job satisfaction, we can use the formula:
ρ = 1 - (6 * Σd^2) / (n * (n^2 - 1))
Where:  
- d is the difference between the ranks of each pair of observations
- n is the number of observations   
Interpretation:
- If ρ is close to +1, it indicates a strong positive monotonic relationship, meaning that as sleep increases, job satisfaction tends to increase.
- If ρ is close to -1, it indicates a strong negative monotonic relationship, meaning that as sleep increases, job satisfaction tends to decrease.
- If ρ is close to 0, it indicates no monotonic relationship between the two variables. 


"""
```
